```python
# firstly import the data into NDarrays
%matplotlib inline
import numpy as np
import pandas as pd
from pandas import read_csv
import matplotlib.pyplot as plt

N1 = np.loadtxt("N1.txt")
N2 = np.loadtxt("N2.txt")
N3 = np.loadtxt("N3.txt")
REM = np.loadtxt("REM.txt")
wake = np.loadtxt("wake.txt")
time = N1[:,0]
N1_signal = N1[:,1]
N2_signal = N2[:,1]
N3_signal = N3[:,1]
REM_signal = REM[:,1]
wake_signal = wake[:,1]
```

# combine the raw signals into one nDarray

```python
Raw_data = np.zeros([N1_signal.shape[0], 5])
Raw_data[:, 0] = N1_signal
Raw_data[:, 1] = N2_signal
Raw_data[:, 2] = N3_signal
Raw_data[:, 3] = REM_signal
Raw_data[:, 4] = wake_signal
```

# now we can obtain the STFT of each window frame for each sleeping stage and structure the input matrix accordingly

```python
# define a function which return a structured input (vectorized STFT image) given all necessary parameters
from scipy import signal
data_stft = np.zeros((735,115*5))
for i in range(0, 5):
    for j in range(0, 115):
        f, t, Zxx = signal.stft(Raw_data[100*j:500 + 100*j, i], fs= 100, window= 'hann', nperseg = 40,
                    noverlap = 25)
        Zxx = Zxx.reshape(-1,)
        Zxx = np.absolute(Zxx)
        data_stft[:,i*115+j] = Zxx
data_stft.shape
```

```python
# now we can create the label
stage_label = np.zeros((575,))
stage_label[0:155] = 1
stage_label[155:310] = 2
stage_label[310:465] = 3
stage_label[465:620] = 4
stage_label[620:775] = 5
stage_label
```

# firstly try a linear SVM classification

```python
data_stft = data_stft.transpose()
data_stft.shape
```

```python
# now we can shuffle the data 
np.random.seed(4)
ran_idx = np.random.permutation(575)
data_stft_shuffled = data_stft[ran_idx]
stage_label_shuffled = stage_label[ran_idx]
```

```python
data_train = data_stft_shuffled[0:450,:]
stage_train = stage_label_shuffled[0:450]
data_test = data_stft_shuffled[450:, :]
stage_test = stage_label_shuffled[450:]
```

```python
from sklearn.svm import LinearSVC
lin_clf = LinearSVC(random_state = 42)
lin_clf.fit(data_train, y=stage_train)
```

```python
from sklearn.metrics import accuracy_score
stage_predict = lin_clf.predict(data_train)
accuracy_score(stage_predict,stage_train)
```

```python
N1_matrix = np.zeros((116, 735))
N2_matrix = np.zeros((116, 735))
N3_matrix = np.zeros((116, 735))
REM_matrix = np.zeros((116, 735))
wake_matrix = np.zeros((116, 735))

# time and frequency resolution tradeoff by choosing differnt segment length
# using overlapping window and overlapping frame break TF representation of each stage into matrixs
for i in range(0,116):
    f,t,Z1 = signal.stft(N1[100*i:500+100*i, 1], fs =100, window ='hann', nperseg = 40, 
                       noverlap = 25)
    f,t,Z2 = signal.stft(N2[100*i:500+100*i, 1], fs =100, window ='hann', nperseg = 40, 
                       noverlap = 25)
    f,t,Z3 = signal.stft(N3[100*i:500+100*i, 1], fs =100, window ='hann', nperseg = 40, 
                       noverlap = 25)
    f,t,Z4 = signal.stft(REM[100*i:500+100*i, 1], fs =100, window ='hann', nperseg = 40, 
                       noverlap = 25)
    f,t,Z5 = signal.stft(wake[100*i:500+100*i, 1], fs =100, window ='hann', nperseg = 40, 
                       noverlap = 25)
    Z1 = np.absolute(Z1)
    Z2 = np.absolute(Z2)
    Z3 = np.absolute(Z3)
    Z4 = np.absolute(Z4)
    Z5 = np.absolute(Z5)
    Z1.astype(np.float16)
    Z2.astype(np.float16)
    Z3.astype(np.float16)
    Z4.astype(np.float16)
    Z5.astype(np.float16)
    Z1 = Z1.reshape((1, Z1.shape[0]*Z1.shape[1]))
    Z2 = Z2.reshape((1, Z2.shape[0]*Z2.shape[1]))
    Z3 = Z3.reshape((1, Z3.shape[0]*Z3.shape[1]))
    Z4 = Z4.reshape((1, Z4.shape[0]*Z4.shape[1]))
    Z5 = Z5.reshape((1, Z5.shape[0]*Z5.shape[1]))
    N1_matrix[i,:] = Z1
    N2_matrix[i,:] = Z2
    N3_matrix[i,:] = Z3
    REM_matrix[i,:] = Z4
    wake_matrix[i,:] = Z5
Z1.shape
```
