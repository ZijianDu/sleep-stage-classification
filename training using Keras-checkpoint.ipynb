```python
# firstly import the data into NDarrays
%matplotlib inline
import numpy as np
import pandas as pd
from pandas import read_csv
import matplotlib.pyplot as plt

N1 = np.loadtxt("N1.txt")
N2 = np.loadtxt("N2.txt")
N3 = np.loadtxt("N3.txt")
REM = np.loadtxt("REM.txt")
wake = np.loadtxt("wake.txt")
time = N1[:,0]
N1_signal = N1[:,1]
N2_signal = N2[:,1]
N3_signal = N3[:,1]
REM_signal = REM[:,1]
wake_signal = wake[:,1]
N1.shape
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
model = Sequential()
model.add(Dense(32, activation='relu', input_dim=4))
model.add(Dense(25, activation = 'relu'))
model.add(Dense(15, activation = 'relu'))
model.add(Dense(5, activation='softmax'))
```

```python
# For a multi-class classification problem
model.compile(optimizer='sgd',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
import numpy as np
# create training input data, which is the original raw signal in this case
data = np.zeros((10000,4))
for i in range (0,4):
    data[0:2000,i] = N1_signal[2000*i:2000*(i+1)]
    data[2000:4000,i] = N2_signal[2000*i:2000*(i+1)]
    data[4000:6000,i] = N3_signal[2000*i:2000*(i+1)]
    data[6000:8000,i] = REM_signal[2000*i:2000*(i+1)]
    data[8000:10000,i] = wake_signal[2000*i:2000*(i+1)]
data.shape
```

```python
# create output label data
labels = np.zeros((10000,5))
#for i in range (0,2):
    # N1 stage
labels[0:2000,0] = 1
    # N2 stage
labels[2000:4000,1] = 1
    # N3 stage
labels[4000:6000,2] = 1
    # REM stage
labels[6000:8000,3] = 1
    # wake stage
labels[8000:10000,4] = 1
Y = labels
#labels
Y
```

```python
# use raw data to train the model, any model optimization and analysis?
history = model.fit(data, Y, epochs=200, batch_size=10)
# list all data in history
#print(history.history.keys())
```

```python
# summarize history for accuracy
fig, ax1 = plt.subplots(figsize=(10,6))
history.history['acc']
t = np.arange(0,200,1)
ax1.plot(t, history.history['acc'], 'b-', alpha = 0.8, label = 'training accuracy')
ax1.set_xlabel("Training epoch", fontsize = 15)
ax1.set_ylabel("Training accuracy", fontsize = 15)
ax2 = ax1.twinx()
ax2.plot(t, history.history['loss'], 'r-', alpha = 0.8, label = 'training loss')
ax2.set_ylabel("Training loss", fontsize = 15)
ax1.legend(loc=2)
ax2.legend(loc = 4)
fig.savefig("RawdataNN.jpg")
```

```python
# second try: using db4 DWT coefficients as input for training
DWT_coeff = pd.read_excel("DWT_coeff.xlsx")
DWT_coeff = DWT_coeff.as_matrix()
```

```python
# split the data 
training_input = np.zeros((25000,8))
#A1
for i in range (0,5):
    training_input[5000*i:5000*(i+1),0] = DWT_coeff[10*i, 0:5000]
#D1
for i in range (0,5):
    training_input[5000*i:5000*(i+1),1] = DWT_coeff[10*i+5, 0:5000]
#A2
for i in range (0,5):
    training_input[5000*i:5000*(i+1),2] = DWT_coeff[10*i+1, 0:5000]
#D2
for i in range (0,5):
    training_input[5000*i:5000*(i+1),3] = DWT_coeff[10*i+6, 0:5000]    
#A3
for i in range (0,5):
    training_input[5000*i:5000*(i+1),4] = DWT_coeff[10*i+2, 0:5000]
#D3
for i in range (0,5):
    training_input[5000*i:5000*(i+1),5] = DWT_coeff[10*i+7, 0:5000]
#A4
for i in range (0,5):
    training_input[5000*i:5000*(i+1),6] = DWT_coeff[10*i+3, 0:5000]
#D4
for i in range (0,5):
    training_input[5000*i:5000*(i+1),7] = DWT_coeff[10*i+8, 0:5000]
    
    
training_input.shape
```

```python
# create the output
training_output = np.zeros((25000, 5))
for i in range (0,5):
    training_output[5000*i:5000*(i+1),i] = 1
training_output.shape
```

```python
# so the training input has 25000x8 dimensions, 25000 are DWT coefficients 
# for five stages of sleep, 5000 points for each stage
# output has 25000x5 dimensions, 25000 corresponds to each row in input
# 5 column corresponds to 5 classes, each stage is labed 1 at corresponding column
import keras
# now build the model #1
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD

model = Sequential()
model.add(Dense(64, input_dim=8, init='uniform'))
model.add(Activation('relu'))
model.add(Dropout(0.1))
model.add(Dense(40, init='uniform'))
model.add(Activation('relu'))

model.add(Dropout(0.1))
model.add(Dense(5, init='uniform'))
model.add(Activation('relu'))
adam = keras.optimizers.Adam(lr=0.002, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
#sgd = SGD(lr=0.03, decay=1e-7, momentum=0.9, nesterov=True)
# compile the model
model.compile(optimizer=adam,
              loss='mse',
              metrics=['accuracy'])
# fit the model
history = model.fit(training_input, training_output, epochs=200, batch_size=30)
score = model.evaluate(training_input, training_output, batch_size=16)
```

```python
fig, ax1 = plt.subplots(figsize=(10,6))
history.history['acc']
t = np.arange(0,200,1)
ax1.plot(t, history.history['acc'], 'b-', alpha = 0.8, label = 'training accuracy')
ax1.set_xlabel("Training epoch", fontsize = 15)
ax1.set_ylabel("Training accuracy",fontsize = 15)
ax2 = ax1.twinx()
ax2.plot(t, history.history['loss'], 'r-', alpha = 0.8, label = 'training loss')
ax2.set_ylabel("Training loss", fontsize = 15)
ax1.legend(loc=2)
ax2.legend(loc = 4)
fig.savefig("DWTcoeffNN.jpg")
```

```python
# now build the model #2
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.optimizers import SGD

model = Sequential()
model.add(Dense(64, input_dim=8, init='uniform'))
model.add(Activation('tanh'))
model.add(Dropout(0.5))
model.add(Dense(64, init='uniform'))
model.add(Activation('tanh'))
model.add(Dropout(0.5))
model.add(Dense(5, init='uniform'))
model.add(Activation('softmax'))

sgd = SGD(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)
# compile the model
model.compile(optimizer=sgd,
              loss='mse',
              metrics=['accuracy'])
# fit the model
model.fit(training_input, training_output, epochs=200, batch_size=20)
score = model.evaluate(training_input, training_output, batch_size=16)
```

```python
# use the DWT coefficients as PCA input to reduce the dimensionality of data
DWT_coeff.shape
N1 = DWT_coeff[0:8, :]
N1 = N1.transpose()
import numpy as np
from sklearn.decomposition import PCA
pca = PCA(n_components = 3, svd_solver = 'full')
pca.fit(N1)
print(pca.explained_variance_ratio_)
N1_pca = pca.fit_transform(N1)
```

```python
# PCA for N2 stage
N2 = DWT_coeff[8:16, :]
N2 = N2.transpose()
pca = PCA(n_components = 3, svd_solver = 'full')
N2_pca = pca.fit_transform(N2)
print(pca.explained_variance_ratio_)
```

```python
# PCA for N3 stage
N3 = DWT_coeff[16:24, :]
N3 = N3.transpose()
pca = PCA(n_components = 3, svd_solver = 'full')
N3_pca = pca.fit_transform(N3)
print(pca.explained_variance_ratio_)
```

```python
# PCA for REM stage
REM = DWT_coeff[24:32, :]
REM = REM.transpose()
pca = PCA(n_components = 3, svd_solver = 'full')
REM_pca = pca.fit_transform(REM)
print(pca.explained_variance_ratio_)
```

```python
# PCA for wake stage
wake = DWT_coeff[32:40, :]
wake = wake.transpose()
pca = PCA(n_components = 3, svd_solver = 'full')
wake_pca = pca.fit_transform(wake)
print(pca.explained_variance_ratio_)
```

```python
# now that we have 3 principle components of DWT coefficients, we can use 
# them to train classifier
# reshape the input matrix
N1_pca.transpose().shape
N1_pca = np.reshape(N1_pca, (1, 3*7681))
N2_pca.transpose().shape
N2_pca = np.reshape(N2_pca, (1, 3*7681))
N3_pca.transpose().shape
N3_pca = np.reshape(N3_pca, (1, 3*7681))
REM_pca.transpose().shape
REM_pca = np.reshape(REM_pca, (1, 3*7681))
wake_pca.transpose().shape
wake_pca = np.reshape(wake_pca, (1, 3*7681))
SVM_input = np.concatenate([N1_pca, N2_pca, N3_pca, REM_pca, wake_pca])
SVM_output = np.arange(1,6)
SVM_output = SVM_output.transpose()
```

```python
# train the SVM for classification
from sklearn.multiclass import OneVsOneClassifier
from sklearn.svm import LinearSVC
classifier = OneVsOneClassifier(LinearSVC(random_state = 0)).fit(SVM_input, SVM_output)
```

```python
# test classification accuracy
import warnings
warnings.filterwarnings('ignore')
prediction = classifier.predict(SVM_input[3,:])
```
